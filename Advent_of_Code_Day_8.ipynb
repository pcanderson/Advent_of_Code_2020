```python
with open('day8_input.txt') as f:
    instructions = f.readlines()
instructions_clean = [instr.strip('\n').split(' ') for instr in instructions]
executions = [0 for _ in range(len(instructions_clean))]

for changed_value in range(0, len(instructions)):
#     print(f"Changed value: {changed_value}")
    instructions2 = [instr.strip('\n').split(' ') for instr in instructions]
    executions2 = [0 for _ in range(len(instructions2))]
    
    if instructions2[changed_value][0] == 'nop':
        instructions2[changed_value][0] = 'jmp'
    elif instructions2[changed_value][0] == 'jmp':
        instructions2[changed_value][0] = 'nop' 
        
    line = 0
    acc = 0
    while True:
#         print(instructions2)
#         print(line)
        if line == len(instructions2):
            print(f"{changed_value} succeeded!")
            print(f"Accumulator: {acc}")
            break
        if line > len(instructions2):
#             print(f"{changed_value} failed (too far: {line})!")
            break
        if max(executions2) > 1:
#             print(f"{changed_value} failed (executions)!")
            break

        if instructions2[line][0] == 'nop':
            executions2[line] += 1
            line += 1
            continue
        if instructions2[line][0] == 'acc':
            acc += int(instructions2[line][1])
            executions2[line] += 1
            line += 1
            continue
        if instructions2[line][0] == 'jmp':
            executions2[line] += 1
            line += int(instructions2[line][1])
            continue
```
